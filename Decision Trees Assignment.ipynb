```python
col_list = dt.columns
col_list
```

```python
for column in col_list:
    if dt[column].dtype!='object':
        sns.boxplot(dt[column])
        plt.xlabel= column
        plt.ylable= ('count')
        plt.show()
```

```python
out_col =["HP","Attack","Defense","Sp. Atk"]
```

```python
for col in out_col:
    Q1=dt[col].quantile(0.25)
    Q3=dt[col].quantile(0.75)
    IQR=Q3-Q1
    lower= Q1-1.5*IQR
    upper=Q3+1.5*IQR
    dt=dt[(dt[col]<=upper)&(dt[col]>=lower)]
```

```python
for column in col_list:
    if dt[column].dtype!='object':
        sns.boxplot(dt[column])
        plt.xlabel= column
        plt.ylable= ('count')
        plt.show()
```

# TEST AND TRAIN DATA

```python
#SPLITTING DATA INTO DEPENDENT AND INDEPENDENT
x=dt.iloc[:,1:-1]
y=dt["Legendary"]
```

```python
dt.head()
```

```python
x=dt.iloc[:,1:-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
y_pred = dt_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import *
```

```python
acc_score = accuracy_score(y_test,y_pred)
```

```python
acc_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```
